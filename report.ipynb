### 2.1 Motion of pendulum
The coupled pendulum has several parameters to explain it's moton during the oscillation.</br>
We define</br>
- length of the pendulum ($L$)
- spring constant ($k$)
- Time period of in-phase oscillation ($T_{in}$)
- Time period pf out-coming oscillation ($T_{out}$)
- Time period of beat mode oscillation ($T$)
- Weght of the mass ($m$)
- Angular displacement from rest position (${\theta}$)

There is one more important value we are looking for which is called "degree of coupling" ($\kappa$).</br>
$$\kappa=\frac{{T_{in}}^2+{T_{out}}^2}{{T_{in}}^2-{T_{out}}^2}=\frac{{{\omega}_{out}}^2-{{\omega}_{in}}^2}{{{\omega}_{out}}^2+{{\omega}_{in}}^2}\tag{1}$$
The value of time period for beat mode oscillation is theoretically calculated with usin this value</br>
We know that force acting on the pendulum of in-phase oscillation is </br>
$$F =-mg\sin(\theta)\tag{2}$$
$g$ = gravitational acceleration</br>
usiing small angle approximation
$$mL\frac{d^2\theta}{dt^2}=-mg\theta\tag{3}$$
This is equal to
$$\ddot{\theta} = -\frac{g}{L}\theta\tag{4}$$
Combine this with general equation of harmonic oscillator
$$\ddot{\theta} = -\omega^2\theta\tag{5}$$
so it gives
$${\omega}_{in}^2 = \frac{g}{L}\tag{6}$$
For out-phase oscillation, there ai additional spring force to in-phase formula
$$mL\frac{d^2\theta}{dt^2}=-mg\theta-2kL\theta\tag{7}$$
$$\ddot{\theta} = -\frac{g}{L}\theta-\frac{2k}{m}\theta\tag{8}$$
Hence, out-phase formula is
$${\omega}_{out}^2 = \frac{g}{L} + \frac{2k}{m}\tag{9}$$
Substitute (6) and (9) into (1)
$$\kappa= \frac{(\frac{g}{L} + \frac{2k}{m})-(\frac{g}{L})}{(\frac{g}{L} + \frac{2k}{m})+(\frac{g}{L})}$$


$$T=\frac{T_{in}T_{out}}{T_{in}-T_{out}}\tag{1}$$
