### 2.1 Motion of pendulum
The coupled pendulum has several parameters to explain it's moton during the oscillation.</br>
We define</br>
- length of the pendulum ($L$)
- spring constant ($k$)
- Time period of in-phase oscillation ($T_{in}$)
- Time period pf out-coming oscillation ($T_{out}$)
- Time period of beat mode oscillation ($T$)
- Weght of the mass ($m$)
- Angular displacement from rest position (${\theta}$)

There is one more important value we are looking for which is called "degree of coupling" ($\kappa$).</br>
$$\kappa=\frac{{T_{in}}^2+{T_{out}}^2}{{T_{in}}^2-{T_{out}}^2}\tag{1}$$
The value of time period for beat mode oscillation is theoretically calculated with
$$T=\frac{T_{in}T_{out}}{T_{in}-T_{out}}\tag{1}$$

Derivation of spring constant $k$ from experimantal values are as follows</br>
Hooke's law
$$F=kx\tag{1}$$
$F$ = forgce exerted on the mass</br>
$x$ = extention of the spring<br>
$k$ = spring constant</br></br>
Force exerted is gravity so that</br>
$$mg=kx\tag{2}$$

$$k=\frac{mg}{x}\tag{3}$$
