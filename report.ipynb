## 1.Introduction
This experiment is to observe the behaviour of two pendulums coupled with spring under some types of oscillations dependent on their initial conditions. Here we use torque constants (denoted as $D$ and $D*$) so that we can include diference in force exerting points smoothly in our calculations. We focus on three types of oscillation mainly two basic motions (called "in-phase oscillation" and "out-phase oscillation") to extend them into complicated motion (called "beat mode oscillation"). Main measurement in our experiment is time period of each oscillations. Property of the spring matters significantly on our observation. Our task is to find out relations of each values with derived theoretical values and compare them with actual experimental values.

## 2.Theoretical Background
### 2.1 Motion of pendulum
The coupled pendulum has several parameters to explain it's moton during the oscillation.</br>
We define</br>
- $L$ - length of the pendulum 
- $s_{A}$ - position of the spring 
- $k$ - spring constant 
- $T_{in}$ - Time period of in-phase oscillation 
- $T_{out}$ - Time period pf out-coming oscillation 
- $T$ - Time period in beat mode oscillation (beat oscillation period)
- $T_s$ - Time period of whole cycle of the beat mode oschillation (beat period)
- $m$ - Weght of the mass 
- $\alpha$ - Initial angle of the pendulum
- $\varphi$ - Angular displacement from initial position 

To explain motion of pendulums, we use torque acting on the pendulum. To simplify them, we define them with single multiplication of constant and angle usin small angle approximation.
- $D$ - Torque constant for the pendulums 
- $D*$ - Torque constant for the pendulums (relating to spring)



### 2.2 Beat oscillation period and beat period

We know that 
$$\frac{1}{T} = \frac{1}{2}\left(\frac{1}{T_{out}}+\frac{1}{T_{in}}\right)\tag{1}$$
for beat oscillation period and
$$\frac{1}{T_s} = \frac{1}{T_{out}}-\frac{1}{T_{in}}\tag{2}$$
for beat period.</br>
These we write in 
$$T=2\left(\frac{T_{out}T_{in}}{T_{in}+T_{out}}\right)\tag{3}$$
and
$$T_s = \frac{T_{out}T_{in}}{T_{in}-T_{out}}\tag{4}$$

### 2.3 Degree of coupling
We derive "degree of coupling" ($\kappa$).</br>
$$\kappa=\frac{T_{in}^2-T_{out}^2}{T_{in}^2+T_{out}^2}\tag{5}$$
This value is used to derive the time period of out-phase oscillation.
The value of time period for beat mode oscillation is theoretically calculated with
$$T=\frac{T_{in}T_{out}}{T_{in}-T_{out}}\tag{6}$$

Derivation of spring constant $k$ from experimantal values are as follows</br>
Hooke's law
$$F=kx\tag{7}$$
$F$ = forgce exerted on the mass</br>
$x$ = extention of the spring<br>
$k$ = spring constant</br></br>
Force exerted is gravity so that</br>
$$k=\frac{mg}{x}\tag{8}$$

So far the calculations shows experimantally measured value of spring constant. Here is one more method to find out the constant but in different method using degree of coupling.
$$\kappa=\frac{ks_A^2}{mgL}\tag{9}$$
$$k=\frac{\kappa mgL}{s_A^2}

### 2.5 influence of the coupling spring position
We use ratio of in-phase against out-phase time period squared.
$$\frac{T_{in}^2}{T_{out}^2}=\frac{2D*+D}{D}\tag{10}$$
